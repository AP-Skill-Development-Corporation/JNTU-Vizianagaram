### Dictionaries

- It a mutable DS with immutable keys and it is ordered.
- we can store any type of data as key:value pairs.
- Represented by {} or dict() method.
- doesn't allow duplicate keys.
- Key&value together call it as an item{key:value}

`````python
student = {1:'abc',2:'xyz',3:'pqr'}
print(student)
`````

`````python
# we can access values by using keys
student[1]
student[2]
`````

`````python
d = {'a':10,'b':20,'b':30,'a':40}
print(d)
`````

`````python
d1 = {'perfect':[6,28],'prime':[2,3,5,7],'set':{2,4,6,8}}
print(d1)
`````

- value may be anything but key must be an immutable type of data.
- ex: value may be a list,tuple,set,dictionary,int,float,str but key must be str or number or tuple only.

`````python
d1['perfect']
`````

`````python
maxx = max(freq.values())
for key,value in freq.items():
    if freq[key]==maxx:
        print(key)
`````

`````python
def is_prime(x):
    for i in range(2,x//2):
        if x%i==0:
            return False
    return True
`````

`````python
is_prime(10000000000000)
`````

`````python
c = 0
n = int(input())
for i in range(2,n):
    if n%i==0:
        c += 1
if c==0:
    print(True)
else:
    print(False)
`````

`````python
for i in range(1,1000000):
    if is_prime(i):
        print(i)
`````

`````python
def is_perfect(n):
    fc=0
    for i in range(1,n):
        if n%i==0:
            fc += i
    if fc==n:
        return True
    else:
        return False
`````

`````python
def even_odd(n):
    if n%2==0:
        return "even"
    else:
        return "Odd"
`````

`````python
def is_palindrome(s):
    if s==s[::-1]:
        return True
    else:
        return False
`````

`````python
def factorial(f):
    if f==0 or f==1:
        return 1
    else:
        return f*factorial(f-1)
`````

`````python
def factors(num):
    for i in range(1,num+1):
        if num%i==0:
            print(i)
`````

`````python
def sqrt(n):
    return n**(0.5)
`````
