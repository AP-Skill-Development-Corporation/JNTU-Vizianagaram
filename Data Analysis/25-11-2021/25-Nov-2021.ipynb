**1. Standard Scaler**
- If dataset is normally distributed and features are not in same scale or contains outliers then, apply standard scaler
- Formula
    - z = ( Xi - mean(X) ) / std(X)

```python
print(df['TV'].mean())
print(df['TV'].std())
```

**Robust Scaler**
- Formula
    - (Xi - median(X) ) / (Q3(X) - Q1(X))
        - Q1 --> Lower Quantile(25%)
        - Q2 --> Median(50%)
        - Q3 --> Upper Quantile(75%)

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
# min values --> LowerQuantile - 1.5 * IQR
# max values --> UpperQuantile + 1.5 * IQR
```

```python
minVal = Q1 - 1.5 * IQR
maxVal = Q3 + 1.5 * IQR
```

```python
df[((df < minVal) | (df > maxVal)).any(axis=1)]
```

```python
df[df > minVal]
```
