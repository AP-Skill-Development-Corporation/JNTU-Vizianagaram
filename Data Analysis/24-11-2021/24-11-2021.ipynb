### Identify duplicates and Null values using Pandas
- To identify duplictaes
    - dataframe.duplicated()
    
- 4 methods to identify null values and manipulate it.
    - isnull()
    - notnull()
    - dropna()
    - fillna()

```python
dict2={'Students':['Mounika','Mounika','Mouni','Sudarsan'],
       'Branch':['MCA','MCA','Mtech','MBA'],
       'ID':[1001,1001,1111,1223]}
df2 = pd.DataFrame(dict2)
df2
```

```python
# to identify duplictes
df2.duplicated()
```

```python
df2[df2.duplicated()]
```

```python
# Remove duplicate values
df2.drop_duplicates()
```

```python
dict3 = {'College':['JNTUA','JNTUK',np.NaN,'JNTUK'],
         'Strength':[5000, np.NaN, 2500, 3890], 
         'Year':[np.NaN, 50, 40, 60],
         'Placed':[4000,3000,2000,np.NaN]}
df2 = pd.DataFrame(dict3)
```

```python
df2
```

```python
# If column is Object type, relpace null value with,
   # Most occured value
# If 2nd col contains null value, replace with before value
# If 3rd col contains null value, replace with next value
# for Remaining null values. replace with '0' 
```

```python
df2['College'].value_counts()
```

```python
df2['College'].fillna('JNTUK', inplace = True)
df2['Strength'].fillna(method = 'ffill', inplace = True)
df2['Year'].fillna(method = 'bfill', inplace = True)
```

```python
df2
```

```python
df2.fillna(0, inplace = True)
```

### Encoding Techniques
- To convert Object type column to Int type column.
    - get_dummies()
    - LabelEncoder()

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic.head()
```

```python
# Check the shape and identify null values in titanic dataset
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
titanic.isnull().sum().sum()
```

```python
pd.get_dummies(titanic['Sex'])
```

```python
titanic['Sex'] = pd.get_dummies(titanic['Sex'])
```

```python
titanic
```

```python
pd.get_dummies(titanic['Embarked'])
```

```python
help(pd.get_dummies)
```

```python
titanic['Embarked'] = pd.get_dummies(titanic['Embarked'])
```

```python
titanic
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
titanic['Sex'] = le.fit_transform(titanic['Sex'])
```

```python
titanic.head()
```

```python
titanic['Embarked'] = le.fit_transform(titanic['Embarked'])
```
