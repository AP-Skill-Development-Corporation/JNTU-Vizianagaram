`````python
product1_sales = np.random.randint(0,5000,50)
product2_sales = np.random.randint(0,7000,100)
product3_sales = np.random.randint(1000,3500,500)
`````

`````python
plt.boxplot([product1_sales, product2_sales, product3_sales],
            labels = ['Maggi','Yeppi','Knoor'],
            patch_artist = True,
            boxprops = dict(facecolor = 'pink',color = 'r'),
            whiskerprops = dict(color = 'green'),
            vert = False)
plt.show()
`````

### 3.Pie Chart
- plt.pie()

`````python
games = ['Cricket','Chess','Ludo','football']
score = [80,92,72,50]
plt.pie(score, labels = games, autopct = '%.f%%',
       colors = ['c','b','m','pink'], startangle = 90,
       explode = (0,0.2,0.05,0), shadow = True)
plt.show()
`````

`````python
games = ['Cricket','Chess','Ludo','football']
score = [80,92,72,50]
plt.pie(score, labels = games, autopct = '%.f%%',
       colors = ['c','b','m','pink'], startangle = 90,
       explode = (0,0.1,0,0), shadow = True)
plt.savefig('PieChart.png')
`````

`````python
games = ['Cricket','Chess','Ludo','football']
score = [80,92,72,50]
plt.pie(score, labels = games, autopct = '%1.2f%%',
        colors = ['c','b','m','pink'], startangle = 90,
        explode = (0,0.2,0,0), shadow = True)
plt.savefig('filename.jpg')
`````

### 4.Area Plot or Stack Plot
- plt.stackplot()

`````python
import numpy as np
import matplotlib.pyplot as plt

x=range(1,6)
y=[1,4,6,8,4]

plt.fill_between(x, y)
plt.show()
`````

`````python
help(plt.stackplot)
`````

`````python
days = [1,2,3,4,5]# X-Scale
work = [16,12,10,11,14]
sleep = [8,10,6,7,8]
# 16+8, 12+10, 10+6, 11+7, 14+8
plt.stackplot(days,work,sleep, colors = ['c','g'],
              labels = ['work','sleep'])
plt.legend(loc = 'center')
plt.show()
`````

`````python
branch = ['cse','ece','mech','eee','civil']
boys = [10,20,25,35,40]
girls = [20,40,45,40,32]
plt.stackplot(branch, boys, girls,
              labels=['boys','girls'])
plt.legend()
plt.savefig('AreaPlot.png')
`````

## Data Visualization using Seaborn

* Seaborn is a Python data visualization library based on matplotlib
* It provides a high-level interface for drawing attractive and informative statistical graphics
* Seaborn is a library for making statistical graphics in Python

**Matplotlib Vs Seaborn**

````
“If Matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.” – Michael Waskom (Creator of Seaborn)
````
