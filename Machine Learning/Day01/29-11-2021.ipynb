#### Linear Regression Model
- Linear regression will ckeck the linear relation between independent and target variables.
    1. Simple Linear Regression
        - Linear relation between one independent variable and dependent varaible
    2. Multiple Linear Regression
        - Linear relation between more than one independent varible and dependent varaible

```python
dict1 = {'Head_Size' : [1,2,3,4,5], 
         'Brain_Weight' : [110,198,285,470,490]}
df = pd.DataFrame(dict1)
df
```

**Line Equation**
- Y = mX + C
- Y = b0 + b1 * X
    - b0 --> Y-Intercept
    - b1 --> Slope

**Slope Formula**

```python
#          (Xi - mean(X)) * (Yi - mean(Y))
#  b1  = ------------------------------------
#              (Xi - mean(X)) ^2
```

```python
X = df['Head_Size'] # Independent or Input
Y = df['Brain_Weight'] # Dependent or Target
```

```python
len(X), len(Y)
```

```python
numerator = 0
denominator = 0
for i in range(len(X)):
    numerator += (X[i] - np.mean(X)) * (Y[i] - np.mean(Y))
    denominator += (X[i] - np.mean(X)) ** 2

b1 = numerator / denominator
print(b1)
```

```python
# b0 = mean(Y) - b1 * mean(X)

b0 = (np.mean(Y) - (b1 * (np.mean(X))))
b0
```

```python
y = b0 + b1 * X
y
```

### Apply Linear Regression Model

**Steps to apply model**
- Step1: Load dataset
- Step2: Divide the columns as Independent and Dependent variables
- Step3: Split the dataset for training and testing
- Step4: Import required model
- Step5: Fit the model and Predict the values
- Step5: Check Accuracy

```python
X1 = df[['Head_Size']]
Y1 = df['Brain_Weight']
```

```python
lr = LinearRegression()
```

```python
lr.fit(X1,Y1)
```

```python
y_predict = lr.predict(X1)
y_predict
```
