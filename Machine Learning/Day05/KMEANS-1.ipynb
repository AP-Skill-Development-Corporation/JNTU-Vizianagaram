## Unsupervised Learning
- It is machine learning technique, where you don't need to supervise the model. Instaed you need to allow the machine to work on its own.
- Clustering: Clustering is the process of dividing the entire data into groups (also known as clusters) based on the patterns in the data
    - K-Means Clustering

### Distance Metrics
- KMeans is a centroid/distance-based algorithm where the distance between each point is calculated and then assigns it to a cluster. The distance can be calculated using
    - Euclidean Distance
    - City block/Manhattan Distance
    
**Euclidean Distance**


## Flow of K-Means Clustering Model



## Updating the centroid

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
d=load_iris()
```

```python
# load data into dataframe
df=pd.DataFrame(d.data,columns=d.feature_names)
df['Target']=d.target
df.sample(5)
```

```python
# select the input data
#x=df.drop('Target',axis=1)
x=df.iloc[:,[0,1,2,3]].values
x
```

```python
# select the algorithm
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(x)
```

```python
y_kmeans = kmeans.predict(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=20, cmap='summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=100, alpha=0.9);
plt.show()
```

## Understanding the Different Evaluation Metrics for Clustering

- Property-1: All the data points in a cluster should be similar to each other
- Property-2: The data points from different clusters should be as different as possible.



## Inertia

- K_means inertia is the measurement of how well the clusters have formed.

- inertia actually calculates the sum of distances of all the points within a cluster from the centroid of that cluster

**Note**: the lesser the inertia value, the better our clusters are



## Dunn Index

- Dunn index is the ratio of the minimum of inter-cluster distances and maximum of intracluster distances.

- The more the value of the Dunn index, the better will be the clusters.



```python
list1 = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(x)
    list1.append(kmeans.inertia_)
plt.plot(range(1,11), list1)
plt.xlabel("K values")
plt.ylabel("Inertia")
plt.show()
```
